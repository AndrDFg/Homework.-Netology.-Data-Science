```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts
%matplotlib inline
```

```python
# @title ПРИМЕР. ПОДСКАЗКА
# @title
rv = sts.norm(loc = 5, scale = 1.5)
sample = rv.rvs(size=1000)
sample[-5:]

# @title
x = np.linspace(0, 10, 1000)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=1)
plt.legend()
plt.ylabel('fraction of samples, f(x)')
plt.xlabel('x')
plt.show()
```

**ДОМАШНЯЯ РАБОТА**

```python
n1 = 30                       # размер выборки
n2 = 60
n3 = 200
n4 = 500
samples_mean1 = []            # листы для хранения средних выборок
samples_mean2 = []
samples_mean3 = []
samples_mean4 = []
samples_count = 1000           # количество генераций
```

```python
rv1 = sts.expon(loc = 1, scale=1.5)
sample = rv1.rvs(1000)
sample[:5]
```

```python
for i in range(samples_count):
  sample_1 = rv1.rvs(n1)
  samples_mean1.append(sample_1.mean())

for i in range(samples_count):
  sample_2 = rv1.rvs(n2)
  samples_mean2.append(sample_2.mean())

for i in range(samples_count):
  sample_3 = rv1.rvs(n3)
  samples_mean3.append(sample_3.mean())

for i in range(samples_count):
  sample_4 = rv1.rvs(n4)
  samples_mean4.append(sample_4.mean())
```

```python
label_1 = 'sample of ' + str(n1) + ' histogram'                      # Подписи для графиков
label_2 = 'sample of ' + str(n2) + ' histogram'
label_3 = 'sample of ' + str(n3) + ' histogram'
label_4 = 'sample of ' + str(n4) + ' histogram'

plt.figure(figsize=(10, 8))
plt.title('Распределение средних')
plt.hist(samples_mean1, bins=30, density=True, label=label_1)
plt.hist(samples_mean2, bins=30, density=True, label=label_2)
plt.hist(samples_mean3, bins=30, density=True, label=label_3)
plt.hist(samples_mean4, bins=30, density=True, label=label_4)        # Строим графики
plt.legend()
plt.ylabel('samples means')
plt.xlabel('x')
```
