**Задание_1**

В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
# Случайная велечина события предприятие - device
device1 = 'событие продукции первого предприятия'
device2 = 'событие продукции второго предприятия'
device3 = 'событии продукция третьего предприятия'
```

```python
# вероятности:
p_device1 = 0.2
p_device2 = 0.3
p_device3 = 0.5
```

```python
# Случайная велечина события сорт - sort
p_device1_sort = 0.1
p_device2_sort = 0.05
p_device3_sort = 0.2
```

```python
res = (p_device1*p_device1_sort) + (p_device2*p_device2_sort) + (p_device3*p_device3_sort)
print(f'Вероятность того, что случайно купленная новая продукция высшего сорта составляет - {res:.2%}')
```

**Задание 3**

Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# формируем данные
data = np.linspace(-4, 4, 100)
```

```python
# Параметры плотности
density_parameters = [(0, 1),(-1, 1),(2, 1), (1, 1), (0, 2), (0, 0.75)]
mu_list = [i[0] for i in density_parameters]
sigma_list = [i[1] for i in density_parameters]

print(f'Параметры loc(среднее значение - μ): {mu_list}')
print(f'Параметры scale(std - σ): {sigma_list}')
```

```python
# Пример со стандартными параметрами
pdf = stats.norm.pdf(data, loc=0, scale=1)
pdf[:10]
```

```python
print(f'Среднее значение - параметр Loc(μ) влияет на смещение графика вправо или влево, распределяет центр.\n'
     f'Стандартное отклонение - параметр scale(σ) влияет на ширину, делает график плоским или остроконечным')

plt.figure(figsize=(12, 7))
for mu, sigma in density_parameters:
  plt.plot(data, stats.norm.pdf(data, mu, sigma), label=f'stats.norm: loc={mu}; scale={sigma}')
plt.title('Нормальное распределение', size='14')
plt.legend()
plt.xlabel('Данные', size='10', fontweight='bold')
plt.ylabel('Плотность - PDF', size='10', fontweight='bold')
plt.show()
```

```python
# @title Отображение отдельными графиками
mas_param = np.array([mu_list, sigma_list])
print(mas_param)

print(f'Параметры loc(среднее значение - μ): {mas_param[0]}')
print(f'Параметры scale(std - σ): {mas_param[1]}')
fig, axs = plt.subplots(2, 3, figsize=(12, 7))
plt.suptitle('Нормальное распределение')
for i in range(2):
  for j in range(3):
    axs[i][j].plot(data, stats.norm.pdf(data, mas_param[0][i*3+j], mas_param[1][i*3+j]), label=f'stats.norm: loc={mas_param[0][i*3+j]}; scale={mas_param[1][i*3+j]}')
    axs[i][j].legend(fontsize=9, loc='center')
```

**Экспоненциальное распределение**

```python
data_expon = np.linspace(0, 8, 100)
expon_lambda = [1, 2, 5, 0.75, 0.25, 0.1]
```

```python
print(f'Уменьшение лямды приводит к увеличению плотности - вероятности, и наоборот увеличение лямды приводит к снижению плотности, вероятность стремится к нулю.\n'
      f'Экспоненциальное распределение имеет только положительные значения, всегда стремится к нулю, но никогда не достигает его.')

fig, axs = plt.subplots(2, 3, figsize=(10, 5))
plt.suptitle('Экспоненциальное распределение')

for i in range(2):
  for j in range(3):
    axs[i, j].plot(data_expon, stats.expon.pdf(data_expon, loc=0, scale=1/expon_lambda[i*3+j]), label=f'lambda={expon_lambda[i*3+j]}')
    axs[i, j].legend()
```

**Распределение Стьюдента**

```python
print(f'Чем больше степень свободы -  nu - параметр df, хвосты графиков уменьшаются и приближаются к нормальнму распределению.\n'
      f'Чем меньше параметр nu, тем больше хвосты графиков, более чувствительны к выбросам.')

data_t_student = np.linspace(-5, 5, 100)
param_df = [1, 5, 10, 100] # nu - степень свободы

fig, axs = plt.subplots(2, 2, figsize=(9, 8))
plt.suptitle('Распределение Стьюдента')

for i in range(2):
  for j in range(2):
    axs[i][j].plot(data_t_student, stats.t.pdf(data_t_student, loc=0, df=param_df[i*2+j]), label=f'nu(df):{param_df[i*2+j]}')
    axs[i][j].legend(loc='upper left', fontsize=10, facecolor='y', edgecolor='b')
```
