**Задание 1**

```python
import numpy as np
import matplotlib.pyplot as plt
import math


A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

**Задача 1**

```python
# Уменьшить объект в два раза.

# Применение диагональной матрицы уменьшения!
matrix_t_1 = np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 0.5]])

A_1 = A @ np.array([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 0.5]])

x = A_1[:,0]
y = A_1[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Диагональная матрица!
matrix_t_1
```

**Задание 2**

Найдите спектральное разложение матрицы:

```python

m = np.array([[1, 2],
              [2, 3]])
print(m)
```

```python
np.linalg.eig(m)
```

```python
# Находим собственные вектора матрицы
vec1 = np.linalg.eig(m)[1][0]
vec2 = np.linalg.eig(m)[1][1]
print(f'Собственные вектора матрицы: {vec1} и {vec2}')
```

```python
# Находим собственные значения матрицы
t1 = np.linalg.eig(m)[0][0]
t2 = np.linalg.eig(m)[0][1]
print(f'Собственные значения матрицы: {t1} и {t2}')
```

```python
# Формируем ортогональную матрицу
Q = np.array([vec1, vec2])
Q
```

```python
Q_T = np.transpose(Q) # ПРОВЕРКА. Получается единичная матрица
Q @ Q_T
```

```python
# Формируем обратную матрицу
Q_inv = np.linalg.inv(Q)
Q_inv
```

```python
Q_T # Транспонированная матрица равна обратной
```

```python
# Формируем диагональную матрицу из собственных значений матрицы m
Q_D = np.diag([t1, t2])
Q_D
```

```python
# матричное переумножение ортогональной, диагональной и обратной(транспонированной) матриц.
A = Q @ Q_D @ Q_inv
A
```
