**'rectal temperature'**

```python
"""По температуре -  медиана, среднее значение, сильно не менялось от того оперировали животное или нет,
а так же погибло животное или нет температура значимо не менялось! Хотя выбросы есть, но такая температура возможна!
Поэтому пустоты заполню по среднему значению всего столбца - rectal temperature
"""
print(F"ВЫБРОСЫ. TЕМП. per rectum: {outlier(horse_data_n['rectal temperature'])}")
horse_data_n.groupby(['surgery?', 'outcome'])['rectal temperature'].median()
#horse_data_n.groupby(['surgery?', 'outcome'])['rectal temperature'].mean()
```
