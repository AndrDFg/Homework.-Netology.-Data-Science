```python
gas_sales_data = pd.read_csv('gas_sales_data.csv')
gas_sales_data.head(1)
```

```python
gas_sales_data.columns
```

```python
# Фильтруем датафрейм. Оставляем только нужные данные
df_south_korea = gas_sales_data[['Seoul', 'Temperature','Sum']]
df_south_korea.info()
```

```python
# Видим высокую (сильную) отрицательную линейную корреляцию между температурой и потреблением газа
corr_ = round(df_south_korea[['Sum', 'Temperature']].corr()['Sum'].values[1], 2)
print(f'Коэффициент корреляции Пирсона: {corr_}')
seaborn.regplot(data=df_south_korea, x='Temperature', y='Sum');
plt.title('Cвязь между температурой и потреблением газа в Южной Корее')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = df_south_korea[['Temperature']]
y = df_south_korea[['Sum']]
```

```python
# random_state - параметр перемешивания
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Размер обучающей выборки (70%)
x_train.shape
```

```python
# Размер обучающей выборки (70%)
y_train.shape
```

```python
# Обучение и подбор оптимальных коэффициентов
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
# Y = b0 + b1X1
# b0 (константа, пересечение)= 71214.3140024
lr_model.coef_
```

```python
# b1 (наклон)= 2622190
# На каждый градус - увеличиение потреблениея газа на сумму 2 622 190
lr_model.intercept_
```

```python
# Тест модели. Ответы Целевого признака на основе x_test
у_prediction = lr_model.predict(x_test)
у_prediction
```

```python
# Правельные ответы
y_test.values
```

**Оценка качества модели**

```python
# Коэффициента детерминации - 73%
# Линейная регрессия соответствует данным, на которых она построена.
# Модель работает хорошо (имеет высокую значимость)
round(lr_model.score(x_test, y_test), 2)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, у_prediction, c='r')
plt.show()
```

```python
# Вычисление остатков
residuals = y_test - у_prediction
residuals
```

```python
plt.scatter(у_prediction, residuals)
#plt.xlabel('Предсказанные значения')
#plt.ylabel('Остатки')
#plt.title('График остатков линейной регрессии')
plt.axhline(y=0, color='blue', linestyle='--')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error as RMSE
```

```python
# Получаем квадрат средней ошибки прогноза выборки теста. Т.Е это модель которая предсказала на основании новых данных
RMSE(y_test, у_prediction)
```

```python
# Получаем квадрат средней ошибки прогноза выборки обучения. Т.Е это модель которая училась на основании предоставленных данных
RMSE(y_train, lr_model.predict(x_train))
```
