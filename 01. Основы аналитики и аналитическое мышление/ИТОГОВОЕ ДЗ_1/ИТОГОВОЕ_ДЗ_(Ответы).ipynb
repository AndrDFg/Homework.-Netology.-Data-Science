**ЗАГРУЗКА СВОЕЙ ТАБЛИЦЫ ОТПУСКОВ**

```python
df_Таблица_Отпусков=pd.read_csv('ИТОГОВОЕ ДЗ (Ответы) - Таблица отпусков.csv',sep=',')
df_Таблица_Отпусков
```

ЗАПИСЬ В ДАТАФРЕЙМ ТАБЛИЦЫ - "City" ИЗ SQL С ФИЛЬТРАЦИЕЙ ПО ГОРОДАМ

```python
df_Таблица_City=pd.read_sql("select*from City WHERE (name='Kazan') OR (name='Moscow') OR (name='Antalya') OR (name='St Petersburg') OR (name='La Habana') OR (name='Budapest')","postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db")
df_Таблица_City
```

**СОЕДИНЕНИЕ ТАБЛИЦ**

```python
df_merge=df_Таблица_Отпусков.merge(df_Таблица_City, left_on='City', right_on='name', how='inner')
df_merge
```

**ИТОГОВЫЙ ДАТАФРЕЙМ ПО СТРАНАМ С РАСЧЁТОМ СРЕДНЕЙ ЧИСЛЕННОСТИ В ГОРОДАХ ГДЕ БЫЛ ОТПУСК!**

---

**Основной, первый вариант!**

```python
df_merge[['Страна','population']].groupby(['Страна']).mean()
```
